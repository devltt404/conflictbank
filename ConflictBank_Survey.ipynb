```python
from datasets import load_dataset
import json
from itertools import islice

# Load dataset with streaming
dataset = load_dataset("Warrieryes/CB_qa", split="train", streaming=True)

subset_size = 5000

with open("cb_qa.jsonl", "w", encoding="utf-8") as f:
    for ex in islice(dataset, subset_size):
        f.write(json.dumps(ex, ensure_ascii=False) + "\n")
```

```python
!mkdir prompts
!python prompt.py --in_file cb_qa.jsonl --out_dir prompts
```

```python
models = ["openai-community/gpt2", "Qwen/Qwen2.5-0.5B", "HuggingFaceTB/SmolLM2-360M"]

# Select models to run inference on
models = models[0:1]
```

```python
from inference import inference

for model in models:
    inference(
        model_name=model,
        input_dir="prompts",
        out_dir=f"runs"
    )
```
